```python
unique_states = df['STATE'].unique()
```

```python
# Create an empty list to store sampled rows
sampled_rows = []
```

```python
# Loop through unique state values and sample one row for each state
for state in unique_states:
    # Sample one row for the current state
    row = df[df['STATE'] == state].sample(1)
      # Append the sampled row to the list
    sampled_rows.append(row)
# Concatenate the sampled rows into a new DataFrame
sampled_df = pd.concat(sampled_rows)
```

```python
# Reset the index of the sampled DataFrame
sampled_df.reset_index(drop=True)
```
